```python
import pandas as pd
import numpy as np
import os
```

```python
#loading channels sheet
from os import listdir

filepaths=["./channel/"+f for f in listdir("./channel/") if f.endswith('.csv')]
rawchan=[]
for filename in filepaths:
    temp=pd.read_csv(filename, header=0, sep="|")
    rawchan.append(temp)
df_channel=pd.concat(rawchan, axis=0, ignore_index=True)
```

```python
#subscription sheets
rawsub=[]
filepaths=["./subscription/"+f for f in listdir("./subscription/") if f.endswith(".csv")]
for filename in filepaths:
    temp=pd.read_csv(filename, header=0, sep="|", error_bad_lines=False)
    rawsub.append(temp)
df_sub=pd.concat(rawsub, axis=0, ignore_index=True)
```

```python
#comment sheets
rawcomment=[]
filepaths=["./comment/"+f for f in listdir("./comment/") if f.endswith(".csv")]
for filename in filepaths:
    temp=pd.read_csv(filename, header=0, sep="|", error_bad_lines=False)
    rawcomment.append(temp)
df_comment=pd.concat(rawcomment, axis=0, ignore_index=True)
```

```python
#video sheets
rawvideo=[]
filepaths=["./video/"+f for f in listdir("./video/") if f.endswith(".csv")]
for filename in filepaths:
    temp=pd.read_csv(filename, header=0, sep="|", error_bad_lines=False)
    rawvideo.append(temp)
df_video=pd.concat(rawvideo, axis=0, ignore_index=True)
```

```python
print(df_channel.columns)
print(df_sub.columns)
print(df_comment.columns)
print(df_video.columns)
```
